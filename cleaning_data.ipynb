```python

# Fill missing values in the original DataFrame as an example
df['reviews_per_month'].fillna(df['reviews_per_month'].mean(), inplace=True)
df['last_review'].fillna(df['last_review'].mode()[0], inplace=True)

# For the provided dataset, here is the code:
df['name'].fillna(df['name'].mode()[0], inplace=True)
df['host_name'].fillna(df['host_name'].mode()[0], inplace=True)
df['last_review'].fillna(df['last_review'].mode()[0], inplace=True)
df['reviews_per_month'].fillna(df['reviews_per_month'].mean(), inplace=True)
```

```python
import matplotlib.pyplot as plt

# Function to detect outliers using the IQR method
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (data < lower_bound) | (data > upper_bound)

# Check for outliers in the 'price' column
outliers_price = detect_outliers_iqr(df['price'])

# Check for outliers in the 'reviews_per_month' column
outliers_reviews_per_month = detect_outliers_iqr(df['reviews_per_month'])

# Plot boxplot for 'price' and 'reviews_per_month' with outliers
fig, axes = plt.subplots(1, 2, figsize=(14, 7))

# Boxplot for 'price'
axes[0].boxplot(df['price'], vert=False, patch_artist=True, showmeans=True)
axes[0].scatter(df[outliers_price]['price'], [1]*outliers_price.sum(), color='red', label='Outliers')
axes[0].set_title('Boxplot of Price')
axes[0].legend()

# Boxplot for 'reviews_per_month'
axes[1].boxplot(df['reviews_per_month'], vert=False, patch_artist=True, showmeans=True)
axes[1].scatter(df[outliers_reviews_per_month]['reviews_per_month'], [1]*outliers_reviews_per_month.sum(), color='red', label='Outliers')
axes[1].set_title('Boxplot of Reviews per Month')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Plot boxplot for 'price' and 'reviews_per_month' without outliers highlighted

fig, axes = plt.subplots(1, 2, figsize=(14, 7))

# Boxplot for 'price'
axes[0].boxplot(df['price'], vert=False, patch_artist=True, showmeans=True)
axes[0].set_title('Boxplot of Price')

# Boxplot for 'reviews_per_month'
axes[1].boxplot(df['reviews_per_month'], vert=False, patch_artist=True, showmeans=True)
axes[1].set_title('Boxplot of Reviews per Month')

plt.tight_layout()
plt.show()

# Check for outliers using the IQR method
outliers_price = detect_outliers_iqr(df['price'])
outliers_reviews_per_month = detect_outliers_iqr(df['reviews_per_month'])

# Number of outliers detected
num_outliers_price = outliers_price.sum()
num_outliers_reviews_per_month = outliers_reviews_per_month.sum()

num_outliers_price, num_outliers_reviews_per_month

```

The boxplots for 'price' and 'reviews_per_month' indicate that there are indeed outliers in both columns. The number of outliers detected are as follows:

Price: 2,972 outliers
Reviews per Month: 3,312 outliers
